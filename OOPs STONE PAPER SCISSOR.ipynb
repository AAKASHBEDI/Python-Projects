```python
import random
Choice_Player=input("Choose b/w Player and Computer : ")
if Choice_Player!="Player" and Choice_Player!="Computer":
    print("Invalid selection of player")
else:   
    No_rounds=int(input("Enter no of rounds you want to play : "))  
    i=1
    while i<=No_rounds:
        print("It is ",i," round", " and ", "No of rounds left are :",No_rounds-i)
        class Computer:
            def __init__(self): 
                pass
            def computer_test(self, computer_choice):
                if computer_choice==1:
                    print("Computer choose Stone")
                    computer_choice="Stone"
                elif computer_choice==2:
                    print("Computer choose Paper")
                    computer_choice="Paper"
                else:
                    print("Computer choose Scissor")
                    computer_choice="Scissor"
                self.computer_choice=computer_choice
                return self.computer_choice

        class Player1:
            def __init__(self):
                pass
            def player1_test(self,player1_choice):
                self.player1_choice=player1_choice
                return self.player1_choice

        class Player2:
            def __init__(self):
                pass
            def player2_test(self,player2_choice):
                self.player2_choice=player2_choice
                return self.player2_choice

        if Choice_Player=="Computer":
            computer_choice=random.randint(1,3)
            c=Computer()
            c.computer_test(computer_choice)
            
            player1_choice=input("Player : Choose b/w Stone,Paper,Scissor\n")
            p1=Player1()
            p1.player1_test(player1_choice)

            if p1.player1_test(player1_choice)!="Stone" and p1.player1_test(player1_choice)!="Paper" and p1.player1_test(player1_choice)!="Scissor":              
                print("Invalid Choice")
            else:
                if p1.player1_test(player1_choice)==c.computer_test(computer_choice):
                    print("It is a Tie")
                else:
                    if p1.player1_test(player1_choice)=="Stone" and c.computer_test(computer_choice)=="Paper": 
                        print("Computer wins!!")
                    elif p1.player1_test(player1_choice)=="Paper" and c.computer_test(computer_choice)=="Scissor":
                        print("Computer wins!!")
                    elif p1.player1_test(player1_choice)=="Scissor" and c.computer_test(computer_choice)=="Stone":
                        print("Computer wins!!")
                    else:
                        print("Player wins!!")
        else:
            player1_choice=input("Player1 : Choose b/w Stone,Paper,Scissor\n")
            p1=Player1()
            p1.player1_test(player1_choice)

            player2_choice=input("Player2 : Choose b/w Stone,Paper,Scissor\n")
            p2=Player2()
            p2.player2_test(player2_choice)

            if (p1.player1_test(player1_choice)!="Stone" and p1.player1_test(player1_choice)!="Paper" and p1.player1_test(player1_choice)!="Scissor") or (p2.player2_test(player2_choice)!="Stone" and p2.player2_test(player2_choice)!="Paper" and p2.player2_test(player2_choice)!="Scissor"):
                print("Invalid Choice")
            else:
                if p1.player1_test(player1_choice)==p2.player2_test(player2_choice):
                    print("It is Tie")
                else:
                    if p1.player1_test(player1_choice)=="Stone" and p2.player2_test(player2_choice)=="Paper":
                        print("Player2 wins!!")
                    elif p1.player1_test(player1_choice)=="Paper" and p2.player2_test(player2_choice)=="Scissor":
                        print("Player2 wins!!")
                    elif p1.player1_test(player1_choice)=="Scissor" and p2.player2_test(player2_choice)=="Stone":
                        print("Player2 wins!!")
                    else:
                        print("Player1 wins!!")
        i=i+1
```
