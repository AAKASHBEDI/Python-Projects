```python
import random
Choice_Player=input("Choose B/w Player and Computer : ")
if Choice_Player!="Player" and Choice_Player!="Computer":
    print("Invalid selection of player")
else:   
    No_rounds=int(input("Enter no of rounds you want to play : "))
    i=0
    while i<No_rounds:
        if Choice_Player=="Computer":
            print("It is Computer Vs. Player")
            print("It is round : ",i+1)
            print("Rounds remaining after this round are : ",No_rounds-(i+1))
            Computer=random.randint(1,3) #Both No are included
            if Computer==1:
                print("Computer choose Stone")
                Computer="Stone"
            elif Computer==2:
                print("Computer choose Paper")
                Computer="Paper"
            else:
                print("Computer choose Scissor")
                Computer="Scissor"
            Player=input("Choose b/w Stone,Paper,Scissor\n")
            if Player!="Stone" and Player!="Paper" and Player!="Scissor" :
                print("Invalid Choice")
            else:
                if Player==Computer:
                    print("It is Tie!!")
                else:
                    if Player=="Stone" and Computer=="Paper":
                        print("Computer wins!!")
                    elif Player=="Paper" and Computer=="Scissor":
                        print("Computer wins!!")
                    elif Player=="Scissor" and Computer=="Stone":
                        print("Computer wins!!")
                    else:
                        print("Player wins!!")
        else:

            print("It is Player1 Vs. Player2")
            print("It is round : ",i+1)
            print("Rounds remaining after this round are : ",No_rounds-(i+1))
            Player1=input("Player1 : Choose b/w Stone,Paper,Scissor\n")
            Player2=input("Player2 : Choose b/w Stone,Paper,Scissor\n")
            if (Player1!="Stone" and Player1!="Paper" and Player1!="Scissor") or (Player2!="Stone" and Player2!="Paper" and Player2!="Scissor"):
                print("Invalid Choice")
            else:
                if Player1==Player2:
                    print("It is Tie!!")
                else:
                    if Player1=="Stone" and Player2=="Paper":
                        print("Player2 wins!!")
                    elif Player1=="Paper" and Player2=="Scissor":
                        print("Player2 wins!!")
                    elif Player1=="Scissor" and Player2=="Stone":
                        print("Player2 wins!!")
                    else:
                        print("Player1 wins!!")

        i=i+1
    print("Game Over")
```
